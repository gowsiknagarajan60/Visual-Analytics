```python
# Step 1: Original poor visualization
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

categories = ['A', 'B', 'C', 'D']
values = [95, 98, 100, 97]

plt.figure(figsize=(8,5))
plt.bar(categories, values, color=['red', 'green', 'blue', 'yellow'])
plt.ylim(90, 105)  # misleading y-axis start
plt.title("Poor Visualization: Truncated Y-axis Bar Chart")
plt.show()

# Step 3: Redesigned visualization
df = pd.DataFrame({'Category': categories, 'Value': values})

plt.figure(figsize=(8,5))
sns.barplot(x='Category', y='Value', data=df, palette='Blues_d')
plt.ylim(0, 110)
plt.title("Redesigned Visualization: Accurate Bar Chart")
plt.xlabel("Category")
plt.ylabel("Value")
for index, row in df.iterrows():
    plt.text(index, row.Value + 1, f'{row.Value}', ha='center')
plt.show()
```
